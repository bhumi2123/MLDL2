```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode categorical variables
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))
    label_encoders[column] = le
```

```python
# Define the column partitions with overlapping columns
part1 = df[['buying', 'persons', 'lug_boot', 'safety', 'class']]
part2 = df[['buying', 'maintainence', 'doors', 'safety', 'class']]
part3 = df[['buying', 'doors', 'persons', 'lug_boot', 'class']]
part4 = df[['buying', 'maintainence', 'doors', 'persons', 'class']]
part5 = df[['buying', 'maintainence', 'doors', 'persons', 'lug_boot', 'class']]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
def prepare_data(df):
    X = df.drop('class', axis=1)
    y = df['class']
    return train_test_split(X, y, test_size=0.3, random_state=42)

X_train1, X_test1, y_train1, y_test1 = prepare_data(part1)
X_train2, X_test2, y_train2, y_test2 = prepare_data(part2)
X_train3, X_test3, y_train3, y_test3 = prepare_data(part3)
X_train4, X_test4, y_train4, y_test4 = prepare_data(part4)
X_train5, X_test5, y_train5, y_test5 = prepare_data(part5)
```

```python
def train_tree(X_train, y_train):
    tree = DecisionTreeClassifier(random_state=42)
    tree.fit(X_train, y_train)
    return tree

tree1 = train_tree(X_train1, y_train1)
tree2 = train_tree(X_train2, y_train2)
tree3 = train_tree(X_train3, y_train3)
tree4 = train_tree(X_train4, y_train4)
tree5 = train_tree(X_train5, y_train5)
```

```python
def predict_with_tree(tree, X_test):
    return tree.predict(X_test)

preds1 = predict_with_tree(tree1, X_test1)
preds2 = predict_with_tree(tree2, X_test2)
preds3 = predict_with_tree(tree3, X_test3)
preds4 = predict_with_tree(tree4, X_test4)
preds5 = predict_with_tree(tree5, X_test5)

accuracy1 = accuracy_score(y_test1, preds1)
accuracy2 = accuracy_score(y_test1, preds2)
accuracy3 = accuracy_score(y_test1, preds3)
accuracy4 = accuracy_score(y_test1, preds4)
accuracy5 = accuracy_score(y_test1, preds5)
print(f'Accuracy of deciosn tree1: {accuracy1}')
print(f'Accuracy of deciosn tree2: {accuracy2}')
print(f'Accuracy of deciosn tree3: {accuracy3}')
print(f'Accuracy of deciosn tree4: {accuracy4}')
print(f'Accuracy of deciosn tree5: {accuracy5}')
```

```python
def majority_vote(*predictions):
    combined = np.array(predictions).T
    final_predictions = []
    for preds in combined:
        final_predictions.append(Counter(preds).most_common(1)[0][0])
    return final_predictions

final_preds = majority_vote(preds1, preds2, preds3, preds4, preds5)
```

```python
accuracy = accuracy_score(y_test1, final_preds)
print(f'Random Forest Accuracy: {accuracy}')
```
