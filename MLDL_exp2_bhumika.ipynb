```python
# Load the datasets
train1 = pd.read_csv('train1_dist.csv')
train2 = pd.read_csv('train2_dist.csv')
train3 = pd.read_csv('train3_dist.csv')
test_data = pd.read_csv('testing_data.csv')

# Define a function to train and evaluate the classifier
def train_and_evaluate(train_data, test_data):
    # Define features and target variable
    X_train = train_data.drop('class', axis=1)
    y_train = train_data['class']
    X_test = test_data.drop('class', axis=1)
    y_test = test_data['class']

    # Initialize and train the Decision Tree Classifier
    clf = DecisionTreeClassifier(random_state=42)
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Train and evaluate classifiers for each training set
accuracy_train1 = train_and_evaluate(train1, test_data)
accuracy_train2 = train_and_evaluate(train2, test_data)
accuracy_train3 = train_and_evaluate(train3, test_data)

# Print the accuracy for each classifier
print(f'Accuracy of classifier trained on train1_dist.csv: {accuracy_train1:.2f}')
print(f'Accuracy of classifier trained on train2_dist.csv: {accuracy_train2:.2f}')
print(f'Accuracy of classifier trained on train3_dist.csv: {accuracy_train3:.2f}')
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from collections import Counter

# Load the datasets
train1 = pd.read_csv('train1_dist.csv')
train2 = pd.read_csv('train2_dist.csv')
train3 = pd.read_csv('train3_dist.csv')
training_data = pd.read_csv('training_data.csv')
test_data = pd.read_csv('testing_data.csv')

# Define a function to train the classifier
def train_classifier(train_data):
    # Define features and target variable
    X_train = train_data.drop('class', axis=1)
    y_train = train_data['class']

    # Initialize and train the Decision Tree Classifier
    clf = DecisionTreeClassifier(random_state=42)
    clf.fit(X_train, y_train)

    return clf

# Train classifiers for each training set
clf1 = train_classifier(train1)
clf2 = train_classifier(train2)
clf3 = train_classifier(train3)
clf_initial = train_classifier(training_data)

# Function to make predictions and perform ensemble voting
def ensemble_predict(sample):
    # Predict with each classifier
    pred1 = clf1.predict([sample])[0]
    pred2 = clf2.predict([sample])[0]
    pred3 = clf3.predict([sample])[0]
    pred_initial = clf_initial.predict([sample])[0]

    # Perform majority voting
    predictions = [pred1, pred2, pred3]
    majority_vote = Counter(predictions).most_common(1)[0][0]

    # Print predictions and the combined result
    print(f'Classifier 1 prediction: {pred1}')
    print(f'Classifier 2 prediction: {pred2}')
    print(f'Classifier 3 prediction: {pred3}')
    print(f'Initial classifier prediction: {pred_initial}')
    print(f'Majority vote prediction: {majority_vote}')

    return pred1, pred2, pred3, pred_initial, majority_vote
```

```python
sample_index = 345
sample = test_data.drop('class', axis=1).iloc[sample_index].values

# Perform ensemble prediction on the sample
ensemble_predict(sample)
```
